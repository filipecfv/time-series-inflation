# Processing the data

```python
import pandas as pd
exchange = pd.read_csv('input/exchange.csv').drop(['RAW DATE', 'CODE', 'DAY', 'MONTH', 'YEAR'], axis = 1)
gdp = pd.read_csv('input/GDP.csv').drop(['RAW DATE', 'CODE', 'DAY', 'MONTH', 'YEAR'], axis = 1)
interest = pd.read_csv('input/interest.csv').drop(['RAW DATE', 'CODE', 'DAY', 'MONTH', 'YEAR'], axis = 1)
ipcar = pd.read_csv('input/IPCA_rate.csv').drop(['RAW DATE', 'CODE', 'DAY', 'MONTH', 'YEAR'], axis = 1)
ipcaa = pd.read_csv('input/IPCA_absolute.csv').drop(['RAW DATE', 'CODE', 'DAY', 'MONTH', 'YEAR'], axis = 1)
oil = pd.read_csv('input/oil_prices.csv').drop(['RAW DATE', 'CODE', 'DAY', 'MONTH', 'YEAR'], axis = 1)
rediscount = pd.read_csv('input/rediscount.csv').drop(['RAW DATE', 'CODE', 'DAY', 'MONTH', 'YEAR'], axis = 1)
```

```python
DF = ipcar.merge(ipcaa, how = 'left', on = 'DATE').rename(columns = {'VALUE ((% a.a.))':'IPCA'})
DF = DF.merge(gdp, how = 'left', on = 'DATE').rename(columns = {'VALUE ((% a.m.))':'D-IPCA', 'VALUE (R$)':'GDP'})
DF = DF.merge(exchange, how = 'left', on = 'DATE').rename(columns = {'VALUE (R$)':'Exchange rate'})
DF = DF.merge(interest, how = 'left', on = 'DATE').rename(columns = {'VALUE ((% a.a.))': 'Interest rate'})
DF = DF.merge(oil, how = 'left', on = 'DATE').rename(columns = {'VALUE (US$)':'Oil price'})
DF = DF.merge(rediscount, how = 'left', on = 'DATE').rename(columns = {'VALUE (R$)':'Rediscount'})
DF
```
